# Weather-prediction Pakistan 2024
The weather forecast for Pakistan in 2024 suggests varying conditions across different regions, according to the Pakistan Meteorological Department (PMD). Here are some key highlights:

# Temperature Trends: 

**Northern Areas:** Expect hot weather, particularly in the plains, with temperatures frequently exceeding 45°C in cities like Lahore, Multan, and Sargodha.

**Coastal Regions: **Karachi and coastal areas will see moderately hot conditions with temperatures around 35-40°C, combined with high humidity levels, making it feel even hotter​ (Pakistan Meteorological Department)​​ (Pakistan Meteorological Department)​.

# Precipitation:
**Monsoon Season**: The monsoon season, typically from June to September, is expected to bring significant rainfall, particularly in northern and northeastern regions. Flash floods and urban flooding are concerns in cities like Karachi, Lahore, and Islamabad during this period.

**Winter Rainfall:** The winter season may bring sporadic rainfall in the northern regions, which can influence agricultural activities and water reservoir levels​ (Pakistan Meteorological Department)​​ (Pakistan Meteorological Department)​
.
# Extreme Weather Events: 
**Heatwaves:** Severe heatwaves are anticipated in many parts of the country, especially in southern Punjab and Sindh, where temperatures could reach up to 50°C.

**Windstorms and Dust Storms:** These are likely in the plains and coastal regions, particularly during the pre-monsoon period (May to June)​ (Pakistan Meteorological Department)​.

# Agricultural Impact:

The seasonal forecasts are crucial for farmers, who rely on these predictions to plan their crops and irrigation. The PMD provides specific advisories to help mitigate the adverse effects of extreme weather conditions on agriculture​ (Pakistan Meteorological Department)​.

Overall, the weather patterns in Pakistan for 2024 will exhibit significant variations, necessitating preparedness for heatwaves, monsoon rains, and occasional storms. For more detailed and updated information, you can visit the Pakistan Meteorological Department's official website​ (Pakistan Meteorological Department)​.

```python
df.drop('month',axis=1,inplace=True)
df.drop('year',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder

df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['weather']=le.fit_transform(df['weather'])
```

```python
x = df[['temp_min', 'temp_max', 'precipitation', 'wind']]
y=df['weather']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

y_pred = nb_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred, zero_division=1)


print(f"Accuracy: {accuracy:.2f}")
```

```python
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(classification_rep)
```
